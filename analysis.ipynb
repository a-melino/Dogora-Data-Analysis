```python
# Import libraries and dependencies
import pandas as pd
import numpy as np
import hvplot.pandas
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import joblib
from pathlib import Path
```

```python
# Import the sales_2022.csv and sales_2023.csv datasets into a Pandas Dataframes
sales_2022_df = pd.read_csv(Path("./Resources/sales_2022.csv"))
sales_2023_df = pd.read_csv(Path("./Resources/sales_2023.csv"))

# Review the DataFrames
display(sales_2022_df.head())
display(sales_2023_df.head())
```

```python
# Drop the unneeded columns from each dataframe
sales_2022_df = sales_2022_df.drop(['POS location', 'Order ID', 'Sale ID', 'Sales channel', 'Product vendor'],axis=1)
sales_2023_df = sales_2023_df.drop(['POS location', 'Order ID', 'Sale ID', 'Sales channel', 'Product vendor'],axis=1)

# Transform the 'Date' columns to a datetime data type
sales_2022_df["Date"] = pd.to_datetime(sales_2022_df["Date"], infer_datetime_format = True, utc = True)
sales_2023_df["Date"] = pd.to_datetime(sales_2023_df["Date"], infer_datetime_format = True, utc = True)

# Review the DataFrames
display(sales_2022_df.head())
display(sales_2023_df.head())
```

```python
sizes = sales_2022_df['Billing country'].value_counts(sort = True)
sizes
```

```python
# Visualize proportion of transactions by country in 2022 by quantity and percentage
billing_country_22 = sns.countplot(x= 'Billing country', data = sales_2022_df)
for container in billing_country_22.containers:
    billing_country_22.bar_label(container)
billing_country_22.set(title = "Transactions by Country - 2022")
plt.savefig('Images/transactions_country_2022.png')

# Visualize proportion of transactions by country in 2022
labels = ['Canada', 'United States', 'Germany', 'United Kingdom']
sizes = sales_2022_df['Billing country'].value_counts(sort=True)
colors = ["red","blue", "yellow", "mediumpurple"]
 
plt.figure(figsize=(7,7))
plt.pie(sizes, explode=(0, 0.025, 0.05, 0.25), labels=labels, colors=colors, autopct='%1.1f%%', labeldistance=1.1)
plt.title('Transactions by Country - 2022')
plt.show()
```

```python
# Visualize proportion of transactions by country in 2023 by quantity and percentage
billing_country_23 = sns.countplot(x= 'Billing country', data = sales_2023_df)
for container in billing_country_23.containers:
    billing_country_23.bar_label(container)
billing_country_23.set(title = "Transactions by Country - 2023")
plt.savefig('Images/transactions_country_2023.png')

# Visualize proportion of transactions by country in 2023
labels = ['Canada', 'United States', 'United Kingdom']
sizes = sales_2023_df['Billing country'].value_counts(sort = True)
colors = ["red","blue", "mediumpurple"]
 
plt.figure(figsize=(7,7))
plt.pie(sizes, explode=(0, 0.05, 0.1), labels=labels, colors=colors, autopct='%1.1f%%', labeldistance=1.1)
plt.title('Transactions by Country - 2023')
plt.show()
```

```python
sales_sums_2022_df = pd.DataFrame(data=(sales_2022_df.sum(numeric_only=True)), columns = ["2022"])
sales_sums_2022_df
```

```python
sales_sums_2023_df = pd.DataFrame(data=(sales_2023_df.sum(numeric_only=True)),columns = ["2023"])
sales_sums_2023_df
```

```python
year_comparison_df = pd.DataFrame(data = (sales_sums_2022_df["2022"], sales_sums_2023_df["2023"]))
year_comparison_df = year_comparison_df.drop(['Net quantity'],axis=1)
year_comparison_df = year_comparison_df.T
year_comparison_df
```

```python
aw = year_comparison_df.plot.bar(
        title="Metric Comparisons 2022 and 2023", xlabel = "", ylabel = "$CAD",
    figsize=[10, 5]
)
aw.figure.savefig('C:/Users/alex_/Desktop/Data Analysis Projects/Dogora - 2022-2023/Dogora-Data-Analysis/Images/metric_comparisons.png')
```

```python
# Distribution of net sales per transactions in the 2022 dataset
histplot2022 = sns.histplot(x = 'Net sales', data = sales_2022_df, bins = 12, binrange = [-600, 600], binwidth = 100)
for container in histplot2022.containers:
    histplot2022.bar_label(container)
histplot2022.set(title="Net sales per Transaction - 2022", xlabel = "Net Sales in $CAD", ylabel = "Transaction Count")
plt.savefig('Images/net_sales_transaction_2022.png')
sales_mean_2022 = sales_2022_df['Net sales'].mean()
print(f"Mean Net Sales per Transaction in 2022: ${sales_mean_2022:.2f}")
```

```python
# Distribution of net sales per transactions in the 2023 dataset
histplot2023 = sns.histplot(x = 'Net sales', data = sales_2023_df, bins = 12, binrange = [-600, 600], binwidth = 100)
for container in histplot2023.containers:
    histplot2023.bar_label(container)
histplot2023.set(title="Net sales per Transaction - 2023", xlabel = "Net Sales in $CAD", ylabel = "Transaction Count")
plt.savefig('Images/net_sales_transaction_2023.png')
sales_mean_2023 = sales_2023_df['Net sales'].mean()
print(f"Mean Net Sales per Transaction in 2023: ${sales_mean_2023:.2f}")

```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
sales_2022_df_monthly = sales_2022_df.set_index("Date")
ay = sales_2022_df_monthly['Total sales'].groupby(by=[sales_2022_df_monthly.index.month]).sum().plot.bar(
        title="Total Sales per Month - 2022", xlabel = "Month", ylabel = "Sales $CAD",
    figsize=[10, 5]
)
ay.figure.savefig('C:/Users/alex_/Desktop/Data Analysis Projects/Dogora - 2022-2023/Dogora-Data-Analysis/Images/sales_2022_monthly.png')
```

```python
sales_2023_df_monthly = sales_2023_df.set_index("Date")
az = sales_2023_df_monthly['Total sales'].groupby(by=[sales_2023_df_monthly.index.month]).sum().plot.bar(
        title="Total Sales per Month - 2023", xlabel = "Month", ylabel = "Sales $CAD",
    figsize=[10, 5]
)
az.figure.savefig('C:/Users/alex_/Desktop/Data Analysis Projects/Dogora - 2022-2023/Dogora-Data-Analysis/Images/sales_2023_monthly.png')
```

```python
sales_total_monthly_df = sales_2022_df_monthly['Total sales'].groupby(by=[sales_2022_df_monthly.index.month]).sum() + sales_2023_df_monthly['Total sales'].groupby(by=[sales_2023_df_monthly.index.month]).sum()

sales_total_monthly_df 
```

```python
ax = sales_total_monthly_df.plot.bar(
        title="Total Sales by Month - 2022-2023", xlabel = "Month", ylabel = "Sales $CAD",
    figsize=[10, 5]
)
ax.figure.savefig('C:/Users/alex_/Desktop/Data Analysis Projects/Dogora - 2022-2023/Dogora-Data-Analysis/Images/sales_total_monthly.png')
```
